```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/data-export.csv')
```

```python
df.head()
```

Data cleaning

```python
# Make 2nd column as header
df.columns = df.iloc[0]
df.head()
```

```python
# Make in Proper Format
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["Channel Group", "DateHour", "Users", "Sessions", "Engaged Sessions", "Average engagement time per session", "Engaged sessions per user", "Events per session", "Engagement rate", "Event count"]
```

```python
df.head()
```

```python
# to check info about the data
df.info()
```

```python
# Convert Object into datetime datatype [DateHour]
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors = 'coerce')
```

```python
# to check
df.info()
```

```python
df.head()
```

```python
# Convert Object datatype to numerical datatype [from users to event count]
numeric_cols = df.columns.drop(["Channel Group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = 'coerce')
# Add a new column by name of Hour
df["Hour"] = df.DateHour.dt.hour
```

```python
# to check
df.info()
```

```python
df.head()
```

```python
# to describe the data
df.describe()
```

Questions Solving

```python
# 1. Sessions and users over time
sns.set(style = "whitegrid")
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over Time")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Total users by channel
plt.figure(figsize=(8,5))
sns.barplot(data = df, x = "Channel Group", y = "Users", estimator= np.sum, palette="viridis")
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Average engagement time by channel
plt.figure(figsize=(8,5))
sns.barplot(data = df, x = "Channel Group", y = "Average engagement time per session", estimator= np.mean, palette="magma")
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Engagement rate distribution by channel
plt.figure(figsize=(8,5))
sns.boxplot(data = df, x = "Channel Group", y = "Engagement rate", palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Engaged vs Nob-engaged sessions
session_df = df.groupby("Channel Group")[["Sessions", "Engaged Sessions"]].sum().reset_index()
session_df["Non-Engaged Sessions"] = session_df["Sessions"] - session_df["Engaged Sessions"]
session_df_melted = session_df.melt(id_vars = "Channel Group", value_vars = ["Engaged Sessions", "Non-Engaged Sessions"])

plt.figure(figsize=(8,5))
sns.barplot(data = session_df_melted, x = "Channel Group", y = "value", hue = "variable", palette="pastel")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=90)
plt.show()
```

```python
# 6. Traffic by hour and channel
heatmap_data = df.groupby(["Hour", "Channel Group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot = True, fmt = '.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of day")
plt.show()
```

```python
# 7. Engagement rate vs sessions over time
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label = "Engagement rate", color = "green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label = "Sessions", color = "blue")
plt.title("Engagement Rate vs Sessions over Time")
plt.xlabel("Date")
plt.legend()
plt.grid = True
plt.show()
```
